```python
import pandas as pd
df = pd.read_csv('COVID clinical trials.csv')
df.head()
```

```python
df.drop(['Results First Posted', 'Study Documents'], axis=1)
```

```python
cat_cols = df.select_dtypes(include='object').columns
for col in cat_cols:
    df[col].fillna(f'Missing {col}')
```

```python
df['Enrollment'].fillna(df['Enrollment'].median())
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract country from Locations column if not already done
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1].strip())

# Top 10 countries
top_countries = df['Country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries by Number of COVID-19 Trials')
plt.xlabel('Number of Trials')
plt.ylabel('Country')
plt.show()
```

```python
status_counts = df['Status'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=status_counts.index, y=status_counts.values)
plt.title('Trial Status Distribution')
plt.ylabel('Count')
plt.show()
```

```python
# Count values in Gender column
gender_counts = df['Gender'].value_counts()

# Plot as line chart
plt.figure(figsize=(8,5))
plt.plot(gender_counts.index, gender_counts.values, marker='o')

plt.title('Gender Distribution in Trials')
plt.xlabel('Gender')
plt.ylabel('Number of Trials')
plt.show()
```

```python
# Ensure 'Start Date' is datetime
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')

# Filter only trials from 2005 onward
df_filtered = df[df['Start Date'].dt.year >= 2005]

# Group by month and count
monthly_counts = df_filtered['Start Date'].dt.to_period('M').value_counts().sort_index()

# Plot
plt.figure(figsize=(12,6))
plt.plot(monthly_counts.index.to_timestamp(), monthly_counts.values, marker='o')
plt.title('COVID-19 Clinical Trials Started (2005 onwards)')
plt.xlabel('Month')
plt.ylabel('Number of Trials')
plt.show()
```

```python
phases = df['Phases'].value_counts()

plt.figure(figsize=(10,6))
sns.barplot(x=phases.index, y=phases.values)
plt.title('Distribution of Clinical Trial Phases')
plt.xlabel('Phase')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
funding = df['Funded Bys'].value_counts().head(5)  # top 5 funders

plt.figure(figsize=(8,8))
plt.pie(funding.values, labels=funding.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Trial Funding Sources')
plt.axis('equal')  # Equal aspect ratio for a perfect circle
plt.show()
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.read_csv("COVID clinical trials.csv")  # replace with your actual CSV
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1].strip())

# Dropdown for country selection
country_dropdown = widgets.Dropdown(
    options=sorted(df['Country'].dropna().unique()),
    description='Country:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

# Plot function
def plot_by_country(selected_country):
    clear_output(wait=True)
    display(country_dropdown)
    
    subset = df[df['Country'] == selected_country]
    status_counts = subset['Status'].value_counts()
    
    plt.figure(figsize=(8, 5))
    status_counts.plot(kind='bar')
    plt.title(f'Trial Status in {selected_country}')
    plt.xlabel('Status')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Interaction
country_dropdown.observe(lambda change: plot_by_country(change['new']), names='value')
display(country_dropdown)
plot_by_country(country_dropdown.value)
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, clear_output
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv("COVID clinical trials.csv")  # Your actual CSV path
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1].strip())
df['Country'] = df['Country'].astype(str).str.strip()
df['Status'] = df['Status'].astype(str).str.strip()

# Create dropdown widget
country_dropdown = widgets.Dropdown(
    options=sorted(df['Country'].dropna().unique()),
    description='Country:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='50%')
)

# Plotting function
def plot_by_country(selected_country):
    clear_output(wait=True)
    display(country_dropdown)

    subset = df[df['Country'] == selected_country]

    if subset.empty:
        print(f"No data for '{selected_country}'")
    else:
        status_counts = subset['Status'].value_counts()
        plt.figure(figsize=(8, 5))
        status_counts.plot(kind='bar')
        plt.title(f'Trial Status in {selected_country}')
        plt.xlabel('Status')
        plt.ylabel('Count')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

# Observe selection
country_dropdown.observe(lambda change: plot_by_country(change['new']), names='value')

# Initial display
display(country_dropdown)
plot_by_country(country_dropdown.value)
```

```python
import dash
from dash import dcc, html, Output, Input
import pandas as pd
import plotly.express as px

# Load data
df = pd.read_csv("COVID clinical trials.csv")
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1].strip())

# Clean-up (optional but helpful)
df['Country'] = df['Country'].astype(str).str.strip()
df['Status'] = df['Status'].astype(str).str.strip()

# Initialize app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    html.H2("COVID Clinical Trials Dashboard", style={'textAlign': 'center'}),
    
    dcc.Dropdown(
        id='country_dropdown',
        options=[{'label': c, 'value': c} for c in sorted(df['Country'].dropna().unique())],
        value='United States',  # default value
        style={'width': '50%'}
    ),
    
    html.Br(),
    
    dcc.Graph(id='status_bar_chart'),
    dcc.Graph(id='funding_pie_chart')
])

# Callbacks for interactivity
@app.callback(
    Output('status_bar_chart', 'figure'),
    Output('funding_pie_chart', 'figure'),
    Input('country_dropdown', 'value')
)
def update_dashboard(selected_country):
    # Filtered data
    filtered_df = df[df['Country'] == selected_country]

    # Bar chart of trial status
    status_fig = px.bar(filtered_df['Status'].value_counts().reset_index(),
                        x='index', y='Status',
                        labels={'index': 'Trial Status', 'Status': 'Count'},
                        title=f"Trial Status in {selected_country}")

    # Pie chart for top 5 funding sources
    funding_fig = px.pie(filtered_df['Funded Bys'].value_counts().head(5).reset_index(),
                         names='index', values='Funded Bys',
                         title=f"Top 5 Funding Sources in {selected_country}",
                         hole=0.4)

    return status_fig, funding_fig

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
